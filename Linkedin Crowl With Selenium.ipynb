```python
####################
# Import Libraries
####################
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from time import sleep
import pandas as pd
import random
import openpyxl
import warnings
warnings.filterwarnings("ignore")
#!pip install chromedriver-binary-auto

####################
# Login LinkedIn
####################
# Enter LinkedIn Login Page
options = Options()
options.headless = False #  If "options.headless =True" then chrome will be hiden
browser = webdriver.Chrome('C:/Users/hasan.kirmizi/Anaconda3/Lib/site-packages/chromedriver_binary/chromedriver.exe', options=options)
browser.get("https://www.linkedin.com")
login=browser.find_element_by_xpath("/html/body/nav/div/a[2]")
login.click()

# Type Username and Pasword and Click login
email=browser.find_element_by_css_selector("#username")
pasword=browser.find_element_by_css_selector("#password")
email.send_keys("********")# Input Username or Mail
pasword.send_keys("********")# Input Paswords
browser.find_element_by_xpath("//*[@id='organic-div']/form/div[3]/button").click()

# Create a list of names who you want to search
candidate_list = ["Hasan K覺rm覺z覺", "Alperen Ercan",]

####################
# Get Contents 
####################

for i in candidate_list: 
    search=browser.find_element_by_xpath("//*[@id='global-nav-typeahead']/input") # Search butonu
    search.send_keys(i)  # type keyword Name in LinkedIn Search bar    
    search.send_keys(Keys.RETURN) # Search
    sleep(2) 
    try:
        link=browser.find_element_by_xpath("//*[@class='app-aware-link']").get_attribute("href")
        print(link)
    except: continue
    try:
        name=browser.find_element_by_xpath("//*[@class='entity-result__title-text t-16']").text.split('\n')[0].strip()
        print(name)
    except: continue
    try:
        title=browser.find_element_by_xpath("//*[@class='entity-result__primary-subtitle t-14 t-black t-normal']").text
        print(title)
    except: continue
    try:
        location=browser.find_element_by_xpath("//*[@class='entity-result__secondary-subtitle t-14 t-normal']").text
        print(location)
    except: continue

    browser.find_element_by_xpath("//*[@id='ember18']").click() # return main page
       
    sleep(2)# 

browser.quit()
```
